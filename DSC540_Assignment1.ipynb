Next step is to create a correct data frame, and edit the data to be used as needed to run a model

```python
data = pd.read_csv("C:/Users/jdrex/OneDrive/Documents/DSC540/Assignment1/housingdata.csv")
data =pd.DataFrame(data)
```

```python
#create y (dependent) variable
y =pd.DataFrame(data['median_house_value'])
y
```

```python
#separate the other attributes from the predicting attribute
x = pd.DataFrame(data.drop('median_house_value',axis=1))
x
```

```python
# handle categorical variable
ocean_proximity=pd.get_dummies(x,drop_first=True)
# dropping extra column
x= x.drop('ocean_proximity',axis=1)
# concatation of independent variables and new cateorical variable.
x=pd.concat([x,ocean_proximity],axis=1)
x
```

```python
#clean up any nan values
x=pd.DataFrame(np.nan_to_num(x))
y=pd.DataFrame(np.nan_to_num(y))
```

```python
#create OLS model
slr_model=sm.OLS(y,x)
slr_reg=slr_model.fit()
```

```python
#summary of OLS model
slr_reg.summary()
```

```python
#create a x variable that is only certain columns
xvar =pd.DataFrame(x[10])
xvar
```

```python
#split x and y into our testing and training data sets
x_train, x_test, y_train, y_test = train_test_split(xvar, y, test_size = 0.2, random_state = 42)
```

```python
y_train
```

```python
#create LR with training data
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#run the model on the test data, and show the predictions, the coefficients, and the intercept
y_prediction =  lr.predict(x_test)
y_prediction
coef=lr.coef_
intercept=lr.intercept_
print("y_prediction:",y_prediction)
print("coefficients:",coef)
print("intercept:",intercept)
```

```python
#plot
import seaborn as sb

sb.scatterplot(x = 'housing_median_age', y = 'median_house_value', data = data, s = 150, alpha = 0.3, edgecolor = 'white')
plt.plot(xvar, coef*xvar + intercept, color = 'r', linewidth = 3)
plt.title('Median House values', fontsize = 16)
plt.ylabel('Median House Value', fontsize = 12)
plt.xlabel('Housing Median Age', fontsize = 12)

#plt.savefig('enginesize_co2_fit.png')
```

```python
#print the r-Squared value
print("R-Squared value:",lr.score(x_test,y_test))
```
